# Урок 2. Анализ данных и проверка статистических гипотез.

### 1. В чём различие между зависимыми и независимыми выборками?

Ззависимые выборки характеризуются тем, что каждому случаю из одной выборки поставлен в соответствие по определенному критерию случай из другой выборки. Например, измерения какого-либо признака до и после экспериментального воздействия.

Независимые выборки характеризуются тем, что вероятность отбора любого случая одной выборки не зависит от отбора любого из случаев другой выборки. Например, выборки из абсолютно разных профессий.

### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Для параметрических критериев данные выборки обладали определенными свойствами, чаще всего необходимым условием является распределение выборки по какому-то закону, например нормальное распределение. Непараметрические критерии не требуют обладания такими свойствами.

Непараметрические критерии наиболее используемы, когда объем выборок мал. Если данных много (например, n > 100), то имеет смысл использовать параметрические критерии, поскольку при увеличении объема выборки выборочные средние подчиняются нормальному закону. Также непараметрические критерии имеют меньшую статистическую мощность, и если важно обнаружить даже слабые отклонения, то не следует их использовать.
