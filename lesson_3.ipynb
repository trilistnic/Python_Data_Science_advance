# Урок 3. Построение модели классификации.

### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

Micro усреднение метрик является по сути метрикой accuracy и отражает долю верхных срабатываний для всех кклассов. Используется когда нет дисбаланса классов.

Macro усреднение метрик рассчитывается, как среднее арифмитическое метрик по всем классам. Используется когда есть дисбаланс классов и важны верные классификации в каждом классе.

Weighted усреднение метрик рассчитывается, как взвешенное среднее арифмитическое метрик по всем классам (с учетом веса каждого класса). Используется когда есть дисбаланс классов и важен вес класса (когда необходимо учитывать важность класса в зависимости от количества объектов с таким классом).

### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

xgboost использует level-wise методику построения деревьев. Дерево решений строится быстрее стандартного алгоритма так как не нужно проверять каждый лист из-за ограничения глубины всего дерева, а отсечение листьев помогает избежать чрезмерной глубины дерева или переобучения. Также используется принцип регуляризации: исключаются ситуации когда в листе недостаточное количество наблюдений.

lightgbm использует leaf-wise методику построения деревьев (не важна глубина построения дерева, а важна информация по каждому листу). Данный алгоритм нарушает распределение наблюдений из-за чего скорость работы существенно меньше, чем у остальных методов. В данном алгоритме в расчетную часть помещаются все наблюдения с большой ошибкой и небольшая часть с маленькой ошибкой. Также данный алгоритм может работать с категориальными признаками.

catboost использует bootstrap или перестановки в своем принципе ordered boosting. На каждой итерации обучение и прогноз происходят на разных выборках. Лучше остальных алгоритмов работает с категориальными признаками. Также данный алгоритм использует принцип симметричных деревьев (дерево строится с уже выбранными параметрами для сплита).
