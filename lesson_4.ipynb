# Урок 4. Оценка и интерпретация полученной модели. 

### 1. Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

Когда дерево решений становится слишком большим, то оно имеет тенденцию к переобучению. Регуляризация с точки зрения дерева решений означает управление ростом дерева. Прирост сложности, оцениваемый ветвлением из вершины, глубиной и числом листьев в поддереве, является штрафным параметром. Параметры регуляризации:

max_depth - ограничение по максимальной глубине дерева;

min_samples_split - ограничение на минимальное количество объектов для разделения;

min_samples_leaf - ограничение на минимальное количество объектов в листе;

max_feature_size -  ограничение на максимальное количество функций, оцениваемых перед разделением для поиска лучшего.

### 2. По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

Признаки для внутренних узлов деревьев выбираются по какому-либо критерию. Для задач классификации этим критерием может быть прирост информации (уменьшение неопределенности  Джини), а для регрессии это снижение дисперсии.Важность признаков определяется атрибутом feature_importances_ и вычисляется, как среднее значение для каждого дерева в ансамбле деревьев для уменьшения неопределенности Джини (или ошибки mse в задачах регрессии).
